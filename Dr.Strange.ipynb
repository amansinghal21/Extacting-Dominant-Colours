```python
from PIL import Image
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python

pil_image = Image.open("dr-strange.jpg")
```

```python
img = np.array(pil_image)
```

```python
plt.imshow(img)
```

```python
img.shape
```

```python
X = img.reshape(-1,3)
```

```python
X.shape
```

```python
model = KMeans(10)
```

```python
model.fit(X)
```

```python
pallets = model.cluster_centers_.astype(np.uint8)
```

```python
plt.imshow(pallets.reshape(1,-1,3))
```

```python
generated_img = np.array([pallets[items] for items in model.labels_])
```

```python
generated_img.shape
```

```python
plt.imshow(generated_img.reshape(img.shape))
```
